## Definitions

* PROBLEM_NAME -> specific air planning problem that is being solved
* SOLUTION_NAME -> name of search algorithm, if prefixed with : then a heuristic function is applied
* ACTIONS_LENGTH -> total number of actions layers required to solve the problem
* EXPANSIONS -> total number of action and literal layers added to the graph
* GOAL_TESTS -> total checks whether current literal layer satisfies goal
* STATES -> total number of possible states identified in the literal layers of the graph
